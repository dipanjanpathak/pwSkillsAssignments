```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('amazon.csv')
```

# Feature Engineering

```
df.sample(5)
```

```
df.columns
```

```
df = df[[  'discounted_price','actual_price', 'discount_percentage', 'rating', 'rating_count','product_name', 'category','review_title',
       'review_content']]
```

```
df.info()
```

```
df['actual_price']
```

```
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False)
df['discounted_price'] = df['discounted_price'].str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False)
df['actual_price'] = df['actual_price'].str.replace(',', '', regex=False)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```
numeric_columns = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```
df.info()
```

```
df.sample(5)
```

```
df.isnull().sum()
```

# Answers

1 .What is the average rating for each product category?

```
avg_ratings = df.groupby('category')['rating'].mean()
print(avg_ratings)
```

2. What are the top rating_count products by category?

```
top_products = df.sort_values(by='rating_count', ascending=False).groupby('category').head(1)
print(top_products[['product_name', 'rating_count']])
```

3. What is the distribution of discounted prices vs. actual prices?

```
import matplotlib.pyplot as plt
plt.scatter(df['actual_price'], df['discounted_price'], alpha=0.5)
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.title('Discounted vs Actual Prices')
plt.show()
```

4. How does the average discount percentage vary across categories?

```
avg_discount = df.groupby('category')['discount_percentage'].mean()
print(avg_discount)
```

5. What are the most popular product names?

```
popular_products = df.sort_values(by='rating_count', ascending=False).head(10)
print(popular_products[['product_name', 'rating_count']])
```

6. What are the most popular product keywords?

```
from collections import Counter
keywords = df['product_name'].str.cat(sep=' ').lower().split()
common_keywords = Counter(keywords).most_common(10)
print(common_keywords)
```

7. What are the most popular product reviews?

```
popular_reviews = df[df['rating'] == 5].sort_values(by='rating_count', ascending=False).head(5)
print(popular_reviews[['review_title', 'review_content']])
```

8. What is the correlation between discounted_price and rating?

```
correlation = df['discounted_price'].corr(df['rating'])
print('Correlation:', correlation)
```

9. What are the Top 5 categories based on the highest ratings?

```
top_categories = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(5)
print(top_categories)
```

# Spotify Data: Solutions

```
df = pd.read_csv('spotify.csv')
```

```
df.head()
```

```
df.info()
```

1. Load the dataframe and ensure data quality by checking for missing values and duplicate rows.

```

print(df.isnull().sum())
df = df.drop_duplicates()
```

2. What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.

```
df['Popularity'].hist(bins=50)
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Popularity Distribution')
plt.show()
```

3. Is there any relationship between the popularity and the duration of tracks?

```

df = df.rename(columns={'Duration (ms)': 'Duration_(ms)'})

```

```
plt.scatter(df['Duration_(ms)'], df['Popularity'], alpha=0.5)
plt.xlabel('Duration_(ms)')
plt.ylabel('Popularity')
plt.title('Popularity vs Duration')
plt.show()
```

4. Which artist has the highest number of tracks in the dataset?

```
track_counts = df['Artist'].value_counts()
print(track_counts.head(1))
```

5. What are the top 5 least popular tracks in the dataset?

```
least_popular = df.nsmallest(5, 'Popularity')
print(least_popular[['Artist', 'Track Name']])
```

6. Among the top 5 most popular artists, which artist has the highest popularity on average?

```
top_artists = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5)
print(top_artists)
```

7. For the top 5 most popular artists, what are their most popular tracks?

```
top_artists_tracks = df[df['Artist'].isin(top_artists.index)].sort_values(by='Popularity', ascending=False).groupby('Artist').head(1)
print(top_artists_tracks[['Artist', 'Track Name']])
```

8. Visualize relationships between multiple numerical variables using a pair plot.

```

sns.pairplot(df[['Popularity', 'Duration_(ms)']])
plt.show()
```

9. Does the duration of tracks vary significantly across different artists?

```
sns.boxplot(x='Artist', y='Duration_(ms)', data=df)
plt.xticks(rotation=90)
plt.show()
```

10. How does the distribution of track popularity vary for different artists?

```
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=90)
plt.show()
```
